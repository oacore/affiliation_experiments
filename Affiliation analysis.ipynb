```python
import re
def clean_domain(x):
    k = re.search('(([A-Za-z0-9-]+)\.)?(([A-Za-z0-9-]+))\.(\w+)$', x)
    if k:
        return k.group()
    else:
        return x


```

```python
core_domains['clean_domain']= core_domains.domain.apply(clean_domain)
```
